```python
# CDC demographics data

df = pd.read_sas('DEMO_L.xpt', format='xport')
df.head()
```

```python
# columns: DMDHHSIZ = number of people in household, INDFMPIR = radio of family income to poverty

demo_selected = df[['SEQN', 'DMDHHSIZ', 'INDFMPIR', 'RIDRETH3']]
demo_selected['RIDRETH3'] = demo_selected['RIDRETH3'].astype(int)
demo_selected['DMDHHSIZ'] = demo_selected['DMDHHSIZ'].astype(int)
demo_selected
```

```python
merged_df = pd.merge(egg_df, demo_selected, on='SEQN', how='inner')

merged_df
```

```python
merged_df = merged_df.rename(columns={
    'DMDHHSIZ': 'Household_Size',
    'INDFMPIR': 'Income_Poverty_Ratio',
    'DR1IFDCD': 'Food',
    'RIDRETH3': 'Race'
})

merged_df
```
