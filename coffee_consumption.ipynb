```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diet.csv')
```

```python
df.info()
```

```python
# columns: DR1IFDCD = type of food, DR1_040Z = Did you eat this meal at home?, DR1FS = Source of food

select_df = df[['SEQN', 'DR1IFDCD', 'DR1_040Z', 'DR1FS']]

select_df
```

```python
select_df["DR1_040Z"] = select_df["DR1_040Z"].replace(1.0, "Yes")
select_df["DR1_040Z"] = select_df["DR1_040Z"].replace(2.0, "No")

select_df["DR1_040Z"]
```

```python
select_df = select_df.rename(columns={"DR1_040Z": "Eat_At_Home", "DR1IFDCD": "Food_Type", "DR1FS": "Source"})

select_df
```

```python
coffee = pd.read_excel('coffee.xlsx')

coffee
```

```python
merged_df = pd.merge(select_df, coffee, on='Food_Type', how='inner')

merged_df
```

```python
merged_df["Eat_At_Home"].value_counts()
```

```python
merged_df["Source"].value_counts()
```

```python
merged_df["Source"] = merged_df["Source"].replace(1.0, "grocery")
merged_df["Source"] = merged_df["Source"].replace(2.0, "restaurant")
merged_df["Source"] = merged_df["Source"].replace(3.0, "fast food")
merged_df["Source"] = merged_df["Source"].replace(15.0, "common pot/snack tray")
merged_df["Source"] = merged_df["Source"].replace(16.0, "from someone else")
merged_df["Source"] = merged_df["Source"].replace(17.0, "mail order")
merged_df["Source"] = merged_df["Source"].replace(27.0, "convenience store")
```

```python
coffee_rank = merged_df["Source"].value_counts().head(7).reset_index()

coffee_rank
```

```python
coffee_rank.to_csv("coffee_rank.csv", index=False)
```
